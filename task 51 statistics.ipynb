```python
#mean
n=int(input())
l=list(map(int,input().split()))
s=sum(l)
mean=s/n
print(mean)
```

```python
#median
n=int(input())
l=list(map(int,input().split()))
l.sort()
if len(l)%2==0:
    a=l[len(l)//2]
    b=l[len(l)//2-1]
    median=(a+b)/2
elif(len(l)%2!=0):
    median=l[len(l)//2]
print(median)
```

```python
#mode
l=list(map(int,input().split()))
counter=0
mode=l[0]
for i in l:
    curr_freq=l.count(i)
    if curr_freq>counter:
        counter=curr_freq
        mode=i
    if len(set(l))==len(l):
        print("no mode")
print(mode)
        
```

```python
#standard deviation
n=int(input())
l=list(map(int,input().split()))
total=sum(l)
mean=total/n
for i in range(len(l)):
    sd=((l[i]-mean)**2)/n
print(sd)
```
